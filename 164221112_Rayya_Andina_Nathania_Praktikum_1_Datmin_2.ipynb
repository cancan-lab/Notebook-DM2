## **Menyimpan gambar Lenna ke JPEG dan melihat perbedaan arraynya**

```python
import cv2

# Baca citra asli
img_ori = cv2.imread('/content/download (1).png')

# Simpan ke format JPEG
cv2.imwrite('/content/lenna_jpeg.jpg', img_ori)

# Baca ulang hasil JPEG
img_jpeg = cv2.imread('/content/lenna_jpeg.jpg')

# Cek nilai array di pixel pertama
print("Pixel asli (PNG)   :", img_ori[0,0])
print("Pixel JPEG hasil   :", img_jpeg[0,0])
```

```python
# ---- membandingkan perbedaan pixel ----
# Hitung selisih absolut antara PNG dan JPEG
diff = cv2.absdiff(img_ori, img_jpeg)

# Cari nilai selisih maksimum, minimum, rata-rata
print("Perbedaan pixel minimum :", diff.min())
print("Perbedaan pixel maksimum:", diff.max())
print("Perbedaan pixel rata-rata:", diff.mean())

# Tampilkan gambar selisih biar kelihatan area mana yang beda
cv2.imwrite('/content/difference.png', diff)
```

```python
# Bandingkan total pixel yang beda
difference = cv2.absdiff(img_ori, img_jpeg)

# Buat mask: True kalau pixel beda, False kalau sama
mask_diff = np.any(difference != 0, axis=-1)

# Hitung jumlah pixel beda dan total pixel
total_pixels = img_ori.shape[0] * img_ori.shape[1]
diff_pixels = np.count_nonzero(mask_diff)

print("Total pixel       :", total_pixels)
print("Pixel yang berbeda:", diff_pixels)
print("Persentase beda   :", (diff_pixels / total_pixels) * 100, "%")
```

**Interpretasi**

Iya, terdapat perbedaan nilai array antara citra asli (PNG) dan citra hasil simpan ke JPEG. Dari hasil percobaan, pixel pertama pada file PNG memiliki nilai [139 176 255], sedangkan pada file JPEG berubah menjadi [138 175 255]. Selain itu, sekitar 60% pixel pada gambar mengalami perbedaan nilai meskipun perubahannya kecil.

Perbedaan ini terjadi karena PNG menggunakan kompresi lossless sehingga data pixel disimpan apa adanya, sedangkan JPEG menggunakan kompresi lossy. Kompresi lossy mengurangi detail halus dan melakukan pendekatan nilai pixel untuk memperkecil ukuran file. Akibatnya, nilai array pada citra JPEG tidak identik dengan citra aslinya meskipun secara visual perbedaannya hampir tidak terlihat.

## **Modifikasi kode a**

### **Frekuensi 400 Hz**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.io import wavfile

# parameter dasar
fs_400 = 44100        # sampling rate (Hz)
f_tone_400 = 400.0    # frekuensi (Hz)
dur_400 = 1.0         # durasi sinyal (detik)

# waktu diskrit
t_400 = np.arange(0.0, dur_400, 1.0/fs_400, dtype=np.float32)

# sinyal sinus 400 Hz
x_400 = np.sin(2*np.pi*f_tone_400*t_400).astype(np.float32)
```

```python
# simpan ke WAV (nama file juga beda)
wavfile.write("tone_400Hz.wav", fs_400, x_400)
```

```python
# info periode & sampel per siklus
period_400 = 1.0 / f_tone_400             # detik/siklus
samp_per_cycle_400 = fs_400 / f_tone_400  # sampel/siklus
print(f"Periode (T): {period_400*1000:.4f} ms")
print(f"Sampel per siklus: {samp_per_cycle_400:.2f}")
```

```python
# zoom 3 ms pertama
n_zoom_400 = int(0.003 * fs_400)
plt.plot(t_400[:n_zoom_400], x_400[:n_zoom_400])
plt.xlabel("detik"); plt.ylabel("amplitudo")
plt.title("Sinyal 400 Hz (zoom 3 ms)")
plt.show()
```

**Interpretasi**

Iya, sesuai output diatas, periode (T) untuk 400 Hz adalah **2.5000 ms** dengan **110.25** sampel per siklus pada sampling rate 44.1 kHz. Dibanding kode (b) yang 200 Hz, sinyal 400 Hz terdengar lebih tinggi dan bentuk gelombangnya lebih rapat karena frekuensi lebih besar sehingga periodenya lebih pendek. Amplitudo tidak diubah, sehingga perbedaan utamanya ada pada tinggi nada (pitch), bukan volume. Keduanya tetap terdengar jelas karena disampling pada 44.1 kHz, sedangkan frekuensinya hanya 400 Hz, jauh di bawah **setengah laju sampling** (sekitar 22.05 kHz), sehingga tidak bermasalah saat direkam.


## **Modifikasi kode b**

### **Amplitudo 50**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.io import wavfile

fs_200 = 44100        # sampling rate
f_200  = 200.0        # frekuensi
A_50   = 50.0         # amplitudo
dur    = 1.0

t_200  = np.arange(0.0, dur, 1.0/fs_200, dtype=np.float32)
x_50   = (A_50 * np.sin(2*np.pi*f_200*t_200)).astype(np.float32)

# info periode & tinggi gelombang
T_200 = 1.0 / f_200                 # detik
peak  = A_50                        # tinggi puncak
p2p   = 2*A_50                      # puncak-ke-lembah

print(f"Periode: {T_200*1000:.4f} ms")
print(f"Tinggi (amplitudo puncak): {peak}")
print(f"Tinggi puncak-ke-lembah: {p2p}")
```

```python
# zoom 20 ms agar bentuknya jelas
n = int(0.02 * fs_200)
plt.plot(t_200[:n], x_50[:n])
plt.xlabel("detik"); plt.ylabel("amplitudo")
plt.title("200 Hz dengan amplitudo 50 (zoom 20 ms)")
plt.show()
```

```python
# simpan  normalisasi ke [-1,1] agar tidak clipping
wavfile.write("tone_200Hz_amp50.wav", fs_200, (x_50/np.max(np.abs(x_50))).astype(np.float32))
```

**Interpretasi**

Berdasarkan output diatas, untuk sinyal 200 Hz dengan amplitudo 50, periodenya tetap **5.0000 ms** (karena periode hanya dipengaruhi frekuensi), sedangkan tinggi gelombang menjadi **50** untuk puncak dan **100** untuk puncak-ke-lembah. Pada grafik terlihat gelombang berulang tiap 5 ms dan nilai amplitudonya berada di kisaran 50 sampai 50. Untuk bunyinya, **tinggi nada (pitch) tetap sama** seperti kode (b) karena frekuensinya tetap 200 Hz. Amplitudo yang lebih besar seharusnya membuat suara **lebih keras**, tetapi pada kode ini sinyal **dinormalisasi** saat disimpan ke file WAV sehingga level akhirnya kembali ke rentang 1 sampai 1. Akibatnya, file yang dihasilkan terdengar **kurang lebih sama kerasnya** dengan file sebelumnya; perbedaannya hanya akan terasa jika disimpan tanpa normalisasi (bisa terdengar lebih keras atau bahkan distorsi/clipping).

## **Modifikasi Kode b**

### **Menampilkan kembali kata pertama yang di lakukanone-hot encoding**

```python
from numpy import array
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
import numpy as np

docs = "I ate an apple"

# memisah kalimat menjadi token
split_docs = docs.split(" ")
data = [doc.split(" ") for doc in split_docs]
values = array(data).ravel()

# integer encode
label_encoder = LabelEncoder()
integer_encoded = label_encoder.fit_transform(values)
print(integer_encoded)

# one-hot encode
onehot_encoder = OneHotEncoder()
integer_encoded = integer_encoded.reshape(len(integer_encoded), 1)
onehot_encoded = onehot_encoder.fit_transform(integer_encoded).toarray()
print(onehot_encoded)

# ---- Tambahan: inverse untuk menampilkan kata pertama dari hasil one-hot ----
first_onehot = onehot_encoded[0].reshape(1, -1)                 # baris pertama: [1 0 0 0]
first_int = onehot_encoder.inverse_transform(first_onehot)[0, 0] # balik ke label integer
first_word = label_encoder.inverse_transform([first_int])[0]     # balik ke token asli
print("Kata pertama (inverse dari one-hot):", first_word)
```

**Interpretasi**

Kode diatas memecah kalimat **I ate an apple** menjadi empat kata, lalu memberi nomor pada tiap kata dengan **urutan alfabet**. Itu sebabnya didapat **\[0 3 1 2]**: kata I jadi 0, ate jadi 3, an jadi 1, dan apple jadi 2; urutannya bukan sesuai posisi di kalimat, tapi sesuai alfabet. Setelah itu, setiap nomor diubah ke **one-hot** sehingga muncul empat vektor dengan 4 kolom (karena ada 4 kata unik). Baris pertama **\[1. 0. 0. 0.]** berarti kata tersebut berada pada indeks 0 dalam daftar kata, yaitu **I**. Ketika baris one-hot pertama dibalik lagi (inverse) ke label lalu ke kata aslinya, hasilnya **I**. Ini menunjukkan proses encode ke decode berjalan benar: kita bisa mengambil kembali kata pertama dari representasi one-hot tanpa kehilangan informasi.


## **File Data Teks**

### **Scrapping data teks**

```python
# !pip install beautifulsoup4 requests

import re, requests
from bs4 import BeautifulSoup

def fetch_page_text(url, timeout=20):
    headers = {
        "User-Agent": ("Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 "
                       "(KHTML, like Gecko) Chrome/120.0 Safari/537.36"),
        "Accept-Language": "id-ID,id;q=0.9,en-US;q=0.8,en;q=0.7",
    }
    s = requests.Session()

    # --- try: versi normal ---
    r = s.get(url, headers=headers, timeout=timeout)
    r.raise_for_status()
    soup = BeautifulSoup(r.text, "html.parser")

    def extract_from_soup(soup_obj):
        # fokus hanya area isi artikel Wikipedia
        content = soup_obj.select_one("#mw-content-text .mw-parser-output") or soup_obj
        paras = []
        for p in content.find_all("p"):
            t = p.get_text(" ", strip=True)
            if t:
                paras.append(t)
        text_local = " ".join(paras)
        # bersihkan catatan referensi [1], [2], ...
        text_local = re.sub(r"\[\d+\]", "", text_local)
        text_local = re.sub(r"\s+", " ", text_local).strip()
        return text_local

    text = extract_from_soup(soup)

    # --- fallback: kalau terlalu pendek & domain wikipedia, coba versi mobile ---
    if len(text.split()) < 200 and "wikipedia.org" in url and ".m." not in url:
        m_url = url.replace("://", "://m.")
        try:
            r2 = s.get(m_url, headers=headers, timeout=timeout)
            r2.raise_for_status()
            soup2 = BeautifulSoup(r2.text, "html.parser")
            text2 = extract_from_soup(soup2)
            if len(text2.split()) > len(text.split()):
                text = text2
        except Exception:
            pass

    return text
```

```python
# ====== ISI URL DI SINI ======
urls = [
    "https://id.wikipedia.org/wiki/Kecerdasan_buatan"
]

# --- ambil teks per URL (pakai fetch_page_text yang sudah kamu definisikan) ---
docs_text = []
for u in urls:
    try:
        txt = fetch_page_text(u)
        docs_text.append(txt)
        print(f"- {u} | {len(txt.split())} kata")
    except Exception as e:
        print(f"! gagal ambil {u}: {e}")

total_words = sum(len(t.split()) for t in docs_text)
print(f"\nTotal kata dari semua sumber: {total_words}")

# --- simpan korpus ke CSV untuk laporan ---
import pandas as pd
df = pd.DataFrame({
    "url": urls,
    "text": docs_text,
    "word_count": [len(t.split()) for t in docs_text]
})
df.to_csv("news_corpus.csv", index=False)
print("Saved: news_data_text.csv")
```

**Interpretasi**

Hasil scraping menunjukkan bahwa teks dari halaman Wikipedia Kecerdasan buatan berhasil diambil dengan baik. Fungsi mengambil isi utama artikel (paragraf di area konten), membuang elemen non-teks seperti script/style, serta membersihkan tanda rujukan seperti [1], sehingga yang tersisa adalah paragraf-paragraf informatif. Jumlah kata yang terkumpul mencapai 1422 kata, jadi sudah melampaui syarat minimal 500 kata untuk analisis. ### **CountVectorizer**

```python
from sklearn.feature_extraction.text import CountVectorizer
import numpy as np, pandas as pd
from IPython.display import display

# (opsional) stopword sederhana bhs Indonesia
stopwords_id = {
    "yang","dan","di","ke","dari","pada","untuk","dengan","sebagai","adalah","itu","ini",
    "atau","juga","karena","oleh","dalam","para","sebuah","seorang","akan","lebih","dapat",
    "saat","antara","maupun","setelah","tanpa","agar","hingga","yakni","bahwa","pun","ia",
    "dia","mereka","kami","kita","saya","menjadi","tersebut","sudah","masih","bisa","serta",
    "namun","yaitu","seperti"
}

pd.set_option("display.max_rows", 50)
pd.set_option("display.max_colwidth", 100)

# ===== Unigram =====
cv_uni = CountVectorizer(ngram_range=(1,1), stop_words=list(stopwords_id))
X_cv_uni = cv_uni.fit_transform(docs_text)
vocab_uni = np.array(cv_uni.get_feature_names_out())
counts_uni = np.asarray(X_cv_uni.sum(axis=0)).ravel()

top_k = 25
top_idx_uni = counts_uni.argsort()[::-1][:top_k]
cv_top_unigram = pd.DataFrame({
    "term": vocab_uni[top_idx_uni],
    "count": counts_uni[top_idx_uni]
})

# simpan & tampilkan
cv_top_unigram.to_csv("cv_top_unigram.csv", index=False)
print("Saved: cv_top_unigram.csv")
print("\nTop Unigram (CountVectorizer):")
display(cv_top_unigram)

# ===== Bigram =====
cv_bi = CountVectorizer(ngram_range=(2,2), stop_words=list(stopwords_id))
X_cv_bi = cv_bi.fit_transform(docs_text)

if X_cv_bi.shape[1] > 0:
    vocab_bi = np.array(cv_bi.get_feature_names_out())
    counts_bi = np.asarray(X_cv_bi.sum(axis=0)).ravel()
    top_idx_bi = counts_bi.argsort()[::-1][:top_k]
    cv_top_bigram = pd.DataFrame({
        "bigram": vocab_bi[top_idx_bi],
        "count": counts_bi[top_idx_bi]
    })
    cv_top_bigram.to_csv("cv_top_bigram.csv", index=False)
    print("\nSaved: cv_top_bigram.csv")
    print("\nTop Bigram (CountVectorizer):")
    display(cv_top_bigram)
else:
    print("\n(Bigram kosong untuk korpus ini)")
```

**Interpretasi**

Hasil CountVectorizer menunjukkan kosakata yang paling sering muncul di artikel Kecerdasan buatan dan cukup menggambarkan topiknya. Pada unigram, kata ai (31), kecerdasan (29), dan buatan (16) mendominasi sehingga menegaskan fokus utama pada AI/Artificial Intelligence. Istilah pendukung seperti komputer, sistem, manusia, mesin, pembelajaran, metode, saraf, jejaring, algoritme, program, dan data ikut sering muncul, yang konsisten dengan bahasan AI sebagai bidang komputasi yang meniru kecerdasan manusia, melibatkan pembelajaran mesin, jaringan saraf tiruan, dan algoritme. Kemunculan kata tahun dan pertama menandakan adanya bagian sejarah atau linimasa perkembangan AI. Masih terlihat kata bahasa Inggris umum seperti the dan an, karena daftar stopword yang dipakai hanya Indonesia.

Pada bigram, frasa kecerdasan buatan (15) paling banyak, selaras dengan unigram kecerdasan dan buatan. Frasa teknis lain seperti jejaring saraf, artificial intelligence, dan intelligence ai ikut muncul, menegaskan pembahasan tentang jaringan saraf dan penyebutan istilah AI dalam dua bahasa. Bigram seperti permainan catur, sistem pakar, perilaku cerdas, dan nama john mccarthy merefleksikan konteks sejarah serta contoh domain AI yang klasik. Ada pula bigram yang kurang informatif yang kemungkinan muncul karena bentuk kalimat tertentu atau pengulangan kata di teks sumberini bisa diminimalkan dengan praproses tambahan (misalnya normalisasi/lemmatization atau filter frasa berfrekuensi sangat rendah).

Secara keseluruhan, frekuensi unigram menegaskan tema besar artikel (AI, sistem, mesin, jaringan saraf, algoritme), sedangkan bigram menangkap istilah majemuk yang lebih spesifik (kecerdasan buatan, jejaring saraf, sistem pakar). Kombinasi keduanya memberikan gambaran yang cukup jelas tentang fokus isi artikel.


### **TF-IDF**

```python
from sklearn.feature_extraction.text import TfidfVectorizer
import numpy as np, pandas as pd
from IPython.display import display

# pakai stopwords yang sama
stopwords_id = {
    "yang","dan","di","ke","dari","pada","untuk","dengan","sebagai","adalah","itu","ini",
    "atau","juga","karena","oleh","dalam","para","sebuah","seorang","akan","lebih","dapat",
    "saat","antara","maupun","setelah","tanpa","agar","hingga","yakni","bahwa","pun","ia",
    "dia","mereka","kami","kita","saya","menjadi","tersebut","sudah","masih","bisa","serta",
    "namun","yaitu","seperti"
}

pd.set_option("display.max_rows", 50)
pd.set_option("display.max_colwidth", 100)

# ===== Unigram =====
tf_uni = TfidfVectorizer(ngram_range=(1,1), stop_words=list(stopwords_id))
X_tf_uni = tf_uni.fit_transform(docs_text)
vocab_uni = np.array(tf_uni.get_feature_names_out())

# kalau hanya 1 dokumen, ambil baris 0
row_uni = X_tf_uni.getrow(0).toarray().ravel()
top_k = 25
top_idx_uni = row_uni.argsort()[::-1][:top_k]

tfidf_top_unigram = pd.DataFrame({
    "term": vocab_uni[top_idx_uni],
    "tfidf": row_uni[top_idx_uni]
})

# simpan & tampilkan
tfidf_top_unigram.to_csv("tfidf_top_unigram.csv", index=False)
print("Saved: tfidf_top_unigram.csv")
print("\nTop Unigram (TF-IDF):")
display(tfidf_top_unigram)

# ===== Bigram =====
tf_bi = TfidfVectorizer(ngram_range=(2,2), stop_words=list(stopwords_id))
X_tf_bi = tf_bi.fit_transform(docs_text)

if X_tf_bi.shape[1] > 0:
    vocab_bi = np.array(tf_bi.get_feature_names_out())
    row_bi = X_tf_bi.getrow(0).toarray().ravel()
    top_idx_bi = row_bi.argsort()[::-1][:top_k]

    tfidf_top_bigram = pd.DataFrame({
        "bigram": vocab_bi[top_idx_bi],
        "tfidf": row_bi[top_idx_bi]
    })
    tfidf_top_bigram.to_csv("tfidf_top_bigram.csv", index=False)
    print("\nSaved: tfidf_top_bigram.csv")
    print("\nTop Bigram (TF-IDF):")
    display(tfidf_top_bigram)
else:
    print("\n(Bigram kosong untuk korpus ini)")
```

**Interpretasi**

Hasil TF-IDF menunjukkan kata dan frasa yang paling penting dalam artikel karena sering muncul dan relevan dengan isi. Di unigram, skor tertinggi ada pada ai (0.451366), diikuti kecerdasan (0.422246), buatan (0.232963), lalu istilah pendukung seperti komputer, sistem, manusia, mesin, pembelajaran, metode, saraf/jejaring, algoritme, program, dan data. Ini sesuai karena artikel memang membahas AI sebagai bidang komputasi yang meniru kecerdasan manusia, dengan topik jaringan saraf, pembelajaran mesin, dan algoritme.

Pada bigram, kecerdasan buatan (0.402911) menjadi frasa paling menonjol,persis judul/topik utama. Muncul juga frasa yang menunjukkan sejarah dan contoh aplikasi seperti permainan catur, sistem pakar, john mccarthy, serta frasa kontekstual seperti pertama kali. Karena korpusnya hanya satu dokumen, nilai IDF untuk semua kata pada dasarnya sama, sehingga peringkat TF-IDF cenderung mengikuti frekuensi yang sudah dinormalisasi (mirip hasil CountVectorizer, hanya skalanya berbeda). ## **Modifikasi Kode d**

### **Menampilkan grafik dari tiap kata**

```python
import matplotlib.pyplot as plt
from sklearn.feature_extraction.text import TfidfVectorizer

text1 = [
    'i love nlp',
    'nlp is so cool',
    'nlp is all about helping machines process language',
    'this tutorial is on basic nlp technique'
]

# fit TF-IDF
tf = TfidfVectorizer()
txt_fitted = tf.fit(text1)

# ambil kata & idf
terms = tf.get_feature_names_out()
idf_vals = txt_fitted.idf_

# (opsional) urutkan dari idf terbesar ke terkecil biar rapi
order = idf_vals.argsort()[::-1]
terms_sorted = terms[order]
idf_sorted = idf_vals[order]

# plot bar chart
plt.figure(figsize=(12,4))
plt.bar(terms_sorted, idf_sorted)
plt.ylabel("IDF")
plt.title("Nilai IDF per kata")
plt.xticks(rotation=60, ha="right")
plt.tight_layout()
plt.show()
```

**Interpretasi**

Grafik IDF di atas nunjukin seberapa spesifik tiap kata di dalam 4 kalimat yang kita pakai. Batang yang tinggi (sekitar **1.9**) berarti kata itu cuma muncul di **1 dokumen** saja, jadi lebih unik dan informatifmisalnya kata seperti *tutorial, technique, language, machines,* dll. Batang setengah tinggi (sekitar **1.22**) itu untuk kata **is** yang muncul di **3 dari 4** kalimat, jadi masih cukup umum. Yang paling pendek (sekitar **1.0**) adalah **nlp**, karena muncul di **semua** kalimat; kata ini dianggap paling umum sehingga bobot IDF-nya paling kecil. Ini sesuai rumus IDF di scikit-learn (versi smoothing), di mana kata yang sering muncul di banyak dokumen akan diturunkan bobotnya, sedangkan kata yang jarang muncul akan dinaikkan bobotnya. Jadi, grafik ini menegaskan bahwa kata-kata spesifik di korpus kita punya IDF tinggi, sedangkan kata yang sangat umum seperti nlp punya IDF rendah.
